# SHAP values tutorial

This tutorial intends to do a small EDA and then look at the feature importance by using the SHAP library. We will use a [Heart Attack Analysis and Prediction](https://www.kaggle.com/rashikrahmanpritom/heart-attack-analysis-prediction-dataset) sampe data set from Kaggle.

This tutorial follows Shap documentation to [build explainable AI models](https://shap.readthedocs.io/en/latest/example_notebooks/overviews/An%20introduction%20to%20explainable%20AI%20with%20Shapley%20values.html).

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import shap
sns.set_style('white')
```

```python
data = pd.read_csv('heart.csv')
```

```python
data.info()
```

**About this dataset**

- Age : Age of the patient
- Sex : Sex of the patient
- exang: exercise induced angina (1 = yes; 0 = no)
- caa: number of major vessels (0-3)
- cp : Chest Pain type chest pain type
    - Value 1: typical angina
    - Value 2: atypical angina
    - Value 3: non-anginal pain
    - Value 4: asymptomatic

- trtbps : resting blood pressure (in mm Hg)
- chol : cholestoral in mg/dl fetched via BMI sensor
- fbs : (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
- restecg : resting electrocardiographic results
    - Value 0: normal
    - Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
    - Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria

- thalach : maximum heart rate achieved
- thall : Thal rate
- **target** : 0= less chance of heart attack 1= more chance of heart attack


```python
data.describe()
```

```python
data.loc[data['sex'] == 0,'sex'] = 'female'
data.loc[data['sex'] == 1,'sex'] = 'male'
```

```python
ax = sns.relplot(x='age', y='trtbps', hue='output', col='sex', data=data)
axes = ax.axes.flatten()
axes[0].set_title('Male')
axes[1].set_title('Female')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,6))
sns.boxplot(x='restecg', y='chol', hue='output', data=data, ax = axes[0])
sns.boxplot(x = 'restecg', y='thalachh', hue='output', data=data, ax=axes[1])
axes[0].set_xlabel('Electrocardiographic Results')
axes[1].set_xlabel('Electrocardiographic Results')
plt.show()
```

```python
X = data.drop('output', axis=1)
y = data['output']
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X, y)
```

We can study the model coefficients to get a sense of how the output would change with respect to the different input variables, however this can be misleading in cases where the coefficients have different scales.

```python
model.coef_
```

To get a better picture we can use the partial dependence plots

```python
shap.plots.partial_dependence("thalachh", model.predict, X, ice=False, model_expected_value=True, feature_expected_value=True)
```

```python
# compute the SHAP values for the linear model
# Xsample = shap.utils.sample(X,300)
explainer = shap.Explainer(model, X)
shap_values = explainer(X)
shap_values.values
```

```python
# make a standard partial dependence plot
sample_ind = 18
shap.partial_dependence_plot(
    "thalachh", model.predict, X, model_expected_value=True,
    feature_expected_value=True, ice=False,
    shap_values=shap_values[sample_ind:sample_ind+1,:]
)
```

```python
X.iloc[sample_ind]['thalachh']
```

```python
shap_values[sample_ind:sample_ind+1,:]
```

```python
shap_values[sample_ind:sample_ind+1,:].values.sum()
```

```python
shap.plots.scatter(shap_values[:,'sex'])
```

```python
# the waterfall_plot shows how we get from shap_values.base_values to model.predict(X)[sample_ind]
shap.plots.waterfall(shap_values[sample_ind], max_display=14)
```

The plot above shows the relevance each of the features have when predicting a single sample (row).
If we want to see how the model behaves on average on the whole data set, we can plot a SHAP bar plot. This takes the mean absolute value of each feature over all the instances (rows) of the dataset.

```python
shap.plots.bar(shap_values)
```

**Now it gets interesting.** The plot we just got is telling us cp, caa, sex variables are among the most significant features.

Recall:

- sex : Sex of the patient
- caa: number of major vessels (0-3)
- cp : Chest Pain type chest pain type
    - Value 1: typical angina
    - Value 2: atypical angina
    - Value 3: non-anginal pain
    - Value 4: asymptomatic
